```python
x = tf.constant(2)
```

```python
y = tf.constant(3)
```

```python
with tf.Session() as sess:
    print("Operations with tensorflow constants")
    print("Addition :", sess.run(x + y))
    print("Subtraction :", sess.run(x - y))
    print("Multiplication :", sess.run(x * y))
    print("Division :", sess.run(x / y))
```

```python
x = tf.placeholder(tf.float64)
y = tf.placeholder(tf.float64)
```

```python
add = tf.add(x, y)
sub = tf.subtract(x, y)
mul = tf.multiply(x, y)
div = tf.divide(x, y)
```

```python
with tf.Session() as sess:
    print("Operations with placeholders")
    print("Addition :", sess.run(add, feed_dict={x:20, y:30}))
    print("Subtraction :", sess.run(sub, feed_dict={x:20, y:30}))
    print("Multiplication :", sess.run(mul, feed_dict={x:20, y:30}))
    print("Division :", sess.run(div, feed_dict={x:20, y:30}))
```

```python
import numpy as np
```

```python
a = np.array([[5.0, 5.0]])
b = np.array([[2.0], [2.0]])
```

```python
a.shape
```

```python
b.shape
```

```python
mat1 = tf.constant(a)
```

```python
type(mat1)
```

```python
mat2 = tf.constant(b)
```

```python
type(mat2)
```

```python
matrix_mult = tf.matmul(mat1, mat2)
```

```python
type(matrix_mult)
```

```python
with tf.Session() as sess:
    result = sess.run(matrix_mult)
    print(result)
```
